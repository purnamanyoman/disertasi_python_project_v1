## 1. Clone repository

```python
from pathlib import Path
import os
import subprocess

REPO_URL = "https://github.com/purnamanyoman/disertasi_python_project_v1.git"
PROJECT_DIR = Path("/content/disertasi_python_project_v1")

if PROJECT_DIR.exists():
    subprocess.run(["git", "-C", str(PROJECT_DIR), "pull"], check=True)
else:
    subprocess.run(["git", "clone", REPO_URL, str(PROJECT_DIR)], check=True)

os.chdir(PROJECT_DIR)
print("Working directory:", Path.cwd())
```

## 5. Upload dataset sendiri

Jalankan cell ini jika ingin memakai dataset CSV dari komputer Anda. Setelah upload selesai, ubah `INPUT_CSV` dan `KEY_COLUMN` pada cell berikutnya.

```python
from google.colab import files

uploaded = files.upload()
for filename, content in uploaded.items():
    target = Path("data/raw") / filename
    target.write_bytes(content)
    print("Uploaded to:", target)
```
